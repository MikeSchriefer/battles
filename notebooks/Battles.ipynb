```python
def enum_to_dict (df):
    return dict(zip(df.value, df.description))
    
```

```python
dflist = [enum_codead, enum_crit, enum_post, enum_post, enum_postype, 
          enum_pri, enum_pri, enum_pri, enum_pri, enum_pri, enum_pri, enum_terra1, enum_terra2, enum_terra3, 
          enum_wina, enum_wx1, enum_wx2, enum_wx3, enum_wx4, enum_wx5, enum_reso, enum_reso, enum_reso]
column_names = ['codead', 'crit', 'post1', 'post2', 'postype', 'pri1', 'pri2', 'pri3', 'sec1', 'sec2', 'sec3',
               'terra1', 'terra2', 'terra3', 'wina', 'wx1', 'wx2', 'wx3', 'wx4', 'wx5', 'reso1', 'reso2', 'reso3']

dict_of_dicts = dict(zip(column_names, [enum_to_dict(df) for df in dflist])) 
#enum_dicts = {} 
#for df in dflist:
   # enum_dicts.update(enum_to_dict(df))
    
```

```python
for col in battles.columns:
    if col in dict_of_dicts:
        battles[col] = battles[col].map(dict_of_dicts[col])
        
```

```python
for col in terrain.columns:
    if col in dict_of_dicts:
        terrain[col] = terrain[col].map(dict_of_dicts[col])
```

```python
for col in weather.columns:
    if col in dict_of_dicts:
        weather[col] = weather[col].map(dict_of_dicts[col])
```

```python
for col in belligerents.columns:
    if col in dict_of_dicts:
        belligerents[col] = belligerents[col].map(dict_of_dicts[col])
```
